```python
Q_EC_76As=0.9233 #in MeV
Z_76As=33

g_a_g_v=1.297 #ratio of vector and axial coupling constants
```

```python
def Q_k_func(E_k, Q_EC):
    #  , E_k -     BGT
    #E_k in KeV
    return Q_EC+E_k/1000
```

   : $\epsilon_e = \frac{E_{\nu}-Q_k}{m_e} + 1$

```python
def e_e_func(E_nu, E_k, Q_EC):
    #    .     E_nu
    # ,  0    
    #E_nu in MeV
    q=Q_k_func(E_k, Q_EC) # 
    if E_nu>q: return 1+(E_nu-q)/0.511
    else: return 0 
```

  : $\pi_e=\sqrt{\epsilon_e^2 -1}$

```python
def Pi_e_func(e_e):
    #  ,    ; **    
    if e_e>1: return math.sqrt(e_e**2 - 1)
    else: return 0
```

```python
def integrate_func(a, b, x, y):
    return np.sum( np.interp(x=np.arange(a, b, 0.001), xp=x, fp=y ) * 0.001 )
```

### Cross-section and total rate

   $\sigma_k={(1.597\times10^{-44}\ cm^2)}\epsilon_e \pi_e m_e^2 F(Z,E_e) [B(F)_k + (\frac{g_a}{g_v})^2 B(GT)_k] $ 

```python
g_a_g_v_square = math.pow(g_a_g_v,2)
```

```python
#     4 
def sigma_k_less4(E_nu, E_k, BGT_k, Z, Q_EC):
    #E_k in kev
    # k- .   
    e_e=e_e_func(E_nu, E_k, Q_EC) #  
    Pi_e=Pi_e_func(e_e) # 
    F_col=num_tables_func(e_e) # -   
    return e_e*Pi_e*F_col*g_a_g_v_square*BGT_k
```

```python
#     4 
def sigma_k_GTR(E_nu, E_k, Z, Q_EC, GTR):
    #WARNING: E_k in KeV 
    e_e=e_e_func(E_nu, E_k, Q_EC) #  
    Pi_e=Pi_e_func(e_e) # 
    F_col=num_tables_func(e_e) # -    
    return e_e*Pi_e*F_col*g_a_g_v_square*GTR
```

```python
def sigma_k_IAS(E_nu, E_k, Z, Q_EC, BF):
    #WARNING: E_k in KeV BUT BF_x, BF_y in MeV
    # k- .   
    e_e=e_e_func(E_nu, E_k, Q_EC) #  
    Pi_e=Pi_e_func(e_e) # 
    F_col=num_tables_func(e_e) # -    
    return e_e*Pi_e*F_col*BF
```

```python
#     
def rate_calc(GTR_x, GTR_y, IAS_x, IAS_y):
# ,   ,  . 
#sigma -     
    E_x_list = np.arange(0,20000, 10)
    #,   -  B(GT)  B(F)  0     
    GTR_list = [integrate_func(0, E_k/1000, GTR_x, GTR_y) for E_k in E_x_list] 
    IAS_list = [integrate_func(0, E_k/1000, IAS_x, IAS_y) for E_k in E_x_list] 
    
    sigma_less4=[sum([sigma_k_less4(E_nu, E_k, BGT_k, Z_76As, Q_EC_76As) 
                      for (BGT_k, E_k) in zip(BGT['B(GT)'].values, BGT['Ex'].values)]) 
                 for E_nu in Energy_all_fluxes]
    
    sigma_GTR=[sum([sigma_k_GTR(E_nu, E_x, Z_76As, Q_EC_76As, GTR) 
                    for (E_x, GTR) in zip(E_x_list, GTR_list)]) 
               for E_nu in Energy_all_fluxes]
    
    sigma_IAS=[sum([sigma_k_IAS(E_nu, E_x, Z_76As, Q_EC_76As, BF) 
                    for (E_x, BF) in zip(E_x_list, IAS_list)]) 
               for E_nu in Energy_all_fluxes]
    
    const = math.pow(0.511,2)*1.597*math.pow(10,-44)
    
    sigma_total = ( np.array(sigma_less4) + np.array(sigma_GTR) + np.array(sigma_IAS) ) * const
    
    for elem in list(BS05_OP_dict.keys()):
        rate = sum(BS05_OP_dict[elem]*sigma_total*all_fluxes_step) * math.pow(10,36)
        print(elem, ' ', rate)
        
#:   pep  
```
